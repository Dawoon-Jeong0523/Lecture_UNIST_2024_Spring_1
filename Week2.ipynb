```python
# 제가 비지니스 현장에 대한 실제 경험은 부족하지만,
# 정책 학술 용역이나 민간 학술 용역을 할 때의 경험에 기반해서 간단한 수요 예측 시뮬레이션을 해보면 아래와 같습니다.

# How many man's coats will be in demand?
# Purchase probability, target height = 178
# alpha는 target height와의 차이
# target height에 딱 맞게 코트를 디자인하면 p만큼 사람들이 구매함
# target height +- alpha를 수요층으로 코트를 디자인하면 코트 구매 가능 대상은 증가하지만 구매 확률은 감소함
# target height에서 키 차이 (alpha) 만큼 구매 확률 p가 감소한다고 가정 ((p*(1-alpha/20)**2))

p = 0.2
temp_list = []
for alpha in range(1,20,1):
  temp_list.append((p*(1-alpha/20)**2)*len(Males[(Males>=178-alpha) & (Males<=178+alpha)]))

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

ax.plot([alpha for alpha in range(1,20,1)], temp_list)
# alpha = 8 정도에서 최대값을 가짐 --> 178cm를 타겟으로 설계하는 코트는 172 ~ 184 사이의 사람들까지 사용가능하게 디자인하는게 예측 구매량이 최대
```

# Binomial distribution and Normal distribution

```python
n, p = 1000000, 0.2
D1 = np.random.binomial(n,p,10000)
D2 = np.random.normal(n*p,np.sqrt(n*p*(1-p)),10000)

plt.hist(D1,bins=100,color='red',label='Binomial')
plt.hist(D2,bins=100,color='blue',label='Normal')
plt.legend()
```

```python
# Large n, binomial --> Normal distribution
s= 1000

fig = plt.figure(figsize=(10,10))

plt.suptitle("Binomial vs. Normal distribution")

ax1 = plt.subplot(221)
n, p = 10, 0.5
D1= np.random.binomial(n,p,s)
D2= np.random.normal(n*p,np.sqrt(n*p*(1-p)),s)
ax1.hist(D1,bins=20,color='red',label='Binomial')
ax1.hist(D2,bins=20,color='blue',label='Normal')
ax1.set_title(f"n = %s"%str(n))
ax1.legend()
ax1.axvline(x=n*p,color='black',linestyle='dotted',linewidth=5)

ax2 = plt.subplot(222)
n, p = 50, 0.5
D1= np.random.binomial(n,p,s)
D2= np.random.normal(n*p,np.sqrt(n*p*(1-p)),s)
ax2.hist(D1,bins=20,color='red',label='Binomial')
ax2.hist(D2,bins=20,color='blue',label='Normal')
ax2.set_title(f"n = %s"%str(n))
ax2.legend()
ax2.axvline(x=n*p,color='black',linestyle='dotted',linewidth=5)

ax3 = plt.subplot(223)
n, p = 100, 0.5
D1= np.random.binomial(n,p,s)
D2= np.random.normal(n*p,np.sqrt(n*p*(1-p)),s)
ax3.hist(D1,bins=20,color='red',label='Binomial')
ax3.hist(D2,bins=20,color='blue',label='Normal')
ax3.set_title(f"n = %s"%str(n))
ax3.legend()
ax3.axvline(x=n*p,color='black',linestyle='dotted',linewidth=5)

ax4 = plt.subplot(224)
n, p = 1000, 0.5
D1= np.random.binomial(n,p,s)
D2= np.random.normal(n*p,np.sqrt(n*p*(1-p)),s)
ax4.hist(D1,bins=20,color='red',label='Binomial')
ax4.hist(D2,bins=20,color='blue',label='Normal')
ax4.set_title(f"n = %s"%str(n))
ax4.legend()
ax4.axvline(x=n*p,color='black',linestyle='dotted',linewidth=5)

plt.tight_layout()
plt.savefig('figure6.jpg',dpi=600)
```

```python
D1.shape, D2.shape
```

```python
D1_quantile = []
D2_quantile = []

for i in range(0,100,1):
  D1_quantile.append(np.quantile(D1,i/100))
  D2_quantile.append(np.quantile(D2,i/100))

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)
ax.scatter(D1_quantile,D2_quantile)
ax.plot([min(D1),max(D1)],[min(D1),max(D1)],color='black',linestyle='dotted')

ax.set_xlabel("Distribution 1",fontsize=20,fontweight='bold')
ax.set_ylabel("Distribution 2",fontsize=20,fontweight='bold')
```

```python
# QQ 플롯은 scipy.stats 라이브러리를 import 하고 probplot 메소드를 활용해서 그릴 수도 있습니다
# 아래는 스스로 만들어본 함수입니다.

def QQ_plot(D1,D2):
  D1_quantile = []
  D2_quantile = []

  for i in range(0,100,1):
    D1_quantile.append(np.quantile(D1,i/100))
    D2_quantile.append(np.quantile(D2,i/100))

  fig = plt.figure(figsize=(10,10))
  ax = plt.subplot(111)
  ax.scatter(D1_quantile,D2_quantile)
  ax.plot([min(D1),max(D1)],[min(D1),max(D1)],color='black',linestyle='dotted')

  ax.set_xlabel("Distribution 1",fontsize=20,fontweight='bold')
  ax.set_ylabel("Distribution 2",fontsize=20,fontweight='bold')
```

```python
D1 = np.random.binomial(10,0.1,10000)
D2 = np.random.binomial(10,0.3,10000)
```

```python
QQ_plot(D1,D2)
```
